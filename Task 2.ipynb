```python
import pandas as pd
```

```python
import zipfile
```

```python
zip_file_path = '/content/titanic.zip'
```

```python
with zipfile.ZipFile(zip_file_path, 'r') as z:
    print("Files in ZIP:", z.namelist())  # Check file names inside ZIP
    with z.open('train.csv') as f:  # Ensure file name matches exactly (case-sensitive)
        df = pd.read_csv(f)
```

```python
# Data Cleaning process
# Lets fill missing age values
df['Age']= df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked']= df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df = df.drop(columns = ['Cabin'])
```

```python
df.isnull().sum()
```

```python
# Performing Exploratory Data Analysis
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data= df, palette= 'pastel')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1= Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# describing gender Vs Survival
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data= df,palette= 'muted')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```

```python
#describing Passenger class VS Survival
plt.figure(figsize=(8,3))
sns. barplot(x='Pclass', hue ='Survived', data=df, palette= 'Set2')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger class', color= 'blue')
plt.ylabel('count', color='red')
plt.show()
```

```python
# Distribution of Age by Survival
plt.figure(figsize=(10,6))
sns.histplot(data= df, x='Age', hue= 'Survived', multiple= 'stack', bins=20, palette= 'coolwarm')
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only= True), annot= True, cmap= 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
